```python
def split2list(df,column,separator=','):
    df[column] = df[column].str.split(separator)
    return df

#find gender in description
def update_column(row):
    male_list=["boy","male","lord"]
    female_list=["girl","female","lady"]
    if row["Gender"] == "" and any(word in row["description"] for word in male_list):
        return "Male"
    else:
        return row["Gender"]

    if row["Gender"] == "" and any(word in row["description"] for word in female_list):
        return "Female"
    else:
        return row["Gender"]

def float2int(row,column):
    #print(row[column])
    if row[column] == "":
        return row[column]
    elif isinstance(row[column], float) and not math.isnan(row[column]):
        return int(row[column])
    elif isinstance(row[column], int) and not math.isnan(row[column]):
        return int(row[column])
    else:
        return "NaN"

```

```python
df["Gender"] = df.apply(update_column, axis=1)
df["Birth year"] = df.apply(float2int, axis=1,column="Birth year")
df["Death year"] = df.apply(float2int, axis=1,column="Death year")
df["Age of death"] = df.apply(float2int, axis=1,column="Age of death")
```
